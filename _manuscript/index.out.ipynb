For this analysis we’ll use the `penguins` dataset from the [**palmerpenguins**](https://allisonhorst.github.io/palmerpenguins/) package.([Gorman, Williams, and Fraser 2014](#ref-gorman2014))

````
library(palmerpenguins)
````

There are 344 in our dataset.

``` python
ggplot(
  penguins,
  aes(
    x = bill_length_mm, y = bill_depth_mm,
    color = species
  )
) +
  geom_point() +
  geom_smooth() +
  scale_color_colorblind() +
  theme(legend.position = "top") +
  labs(x = "Bill length (mm)", y = "Bill depth (mm)")
```
